```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("University_Dataset(tmc_raw_data_2020_2029).csv", parse_dates=["count_date", "start_time", "end_time"])
```

```python
vehicle_columns = [
    "n_appr_cars_r", "n_appr_cars_t", "n_appr_cars_l", "s_appr_cars_r", "s_appr_cars_t", "s_appr_cars_l",
    "e_appr_cars_r", "e_appr_cars_t", "e_appr_cars_l", "w_appr_cars_r", "w_appr_cars_t", "w_appr_cars_l",
    "n_appr_truck_r", "n_appr_truck_t", "n_appr_truck_l", "s_appr_truck_r", "s_appr_truck_t", "s_appr_truck_l",
    "e_appr_truck_r", "e_appr_truck_t", "e_appr_truck_l", "w_appr_truck_r", "w_appr_truck_t", "w_appr_truck_l",
    "n_appr_bus_r", "n_appr_bus_t", "n_appr_bus_l", "s_appr_bus_r", "s_appr_bus_t", "s_appr_bus_l",
    "e_appr_bus_r", "e_appr_bus_t", "e_appr_bus_l", "w_appr_bus_r", "w_appr_bus_t", "w_appr_bus_l"
]

df["total_vehicles"] = df[vehicle_columns].sum(axis=1)
```

## Top 10 Busiest Locations by Total Vehicle Count

```python
location_traffic = df.groupby("location_name")["total_vehicles"].sum().reset_index() # Total number of vehicles in each location
location_traffic = location_traffic.sort_values(by="total_vehicles", ascending=False)

print(location_traffic.head(10))  # Top 10 busiest locations
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=location_traffic.head(10), x="total_vehicles", y="location_name", palette="Blues_r")
plt.xlabel("Total Vehicle Count")
plt.ylabel("Location Name")
plt.title("Top 10 Busiest Locations by Total Vehicle Count")
plt.show()
```

### <span style="color:teal">Insights:</span>

#### <span style="color:teal">Busiest Locations:</span>
The locations with the highest total vehicle counts are concentrated around major streets and intersections, including <span style="color:teal">Mount Pleasant Rd</span>, <span style="color:teal">Yonge St</span>, and <span style="color:teal">Avenue Rd</span>, with counts ranging from <span style="color:teal">294,432 to 443,065</span> vehicles.

#### <span style="color:teal">High Traffic Areas:</span>
These <span style="color:teal">top 10 locations</span> are key traffic hotspots, likely contributing significantly to congestion and delays during peak hours.

#### <span style="color:teal">Traffic Management Potential:</span>
With such high traffic volumes, these locations are ideal candidates for traffic optimization strategies such as <span style="color:teal">traffic light retiming</span>, <span style="color:teal">dedicated bus lanes</span>, and the introduction of <span style="color:teal">sustainable transportation options</span> to improve flow and reduce congestion.

#### <span style="color:teal">Opportunities for Sustainability:</span>
The <span style="color:teal">high volume of vehicles</span> presents a challenge for environmental sustainability, indicating the potential for measures such as increasing <span style="color:teal">public transit accessibility</span> or integrating <span style="color:teal">bike lanes</span> to reduce the environmental impact.


## Locations with the Highest Average Traffic Volume


```python
avg_traffic = df.groupby("location_name")["total_vehicles"].mean().reset_index()
avg_traffic = avg_traffic.sort_values(by="total_vehicles", ascending=False)

print(avg_traffic.head(10))  # Top 10 locations with the highest avg traffic volume
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_traffic.head(10), x="total_vehicles", y="location_name", palette="Reds_r")
plt.xlabel("Average Vehicle Count")
plt.ylabel("Location Name")
plt.title("Top 10 Locations with Highest Average Traffic Volume")
plt.show()
```

### <span style="color:teal">Insights:</span>

The locations with the highest average traffic volume are major arterial roads and highways, including Steeles Ave E, Sheppard Ave E, and William R Allen Rd, with traffic volumes ranging from 1,083 to 1,389 vehicles.

#### <span style="color:teal">Traffic Management:</span>
These areas experience heavy traffic, making them prime candidates for traffic flow optimization initiatives. Potential strategies include improving traffic light timing, creating dedicated bus lanes, and promoting alternate routes to alleviate congestion.

#### <span style="color:teal">Opportunities for Sustainability:</span>
The high traffic volumes emphasize the need for sustainable transportation options such as public transit enhancements, cycling infrastructure, and initiatives to reduce private vehicle usage, which could help mitigate the environmental impact.

#### <span style="color:teal">Urban Planning Focus:</span>
Prioritizing these areas in urban planning efforts is essential, with targeted solutions aimed at reducing congestion and enhancing mobility, making these locations key targets for smart city planning.

## Traffic Volume Across Days of the Week

```python
df["day_of_week"] = df["count_date"].dt.day_name()
daily_traffic = df.groupby("day_of_week")["total_vehicles"].sum().reset_index()

# Ensure correct weekday order
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
daily_traffic["day_of_week"] = pd.Categorical(daily_traffic["day_of_week"], categories=weekday_order, ordered=True)
daily_traffic = daily_traffic.sort_values("day_of_week")

print(daily_traffic)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=daily_traffic, x="day_of_week", y="total_vehicles", palette="coolwarm")
plt.xlabel("Day of the Week")
plt.ylabel("Total Traffic Volume")
plt.title("Traffic Volume Across Days of the Week")
plt.show()
```

### <span style="color:teal">Insights:</span>

- **<span style="color:teal">Highest Traffic Days:</span>**  
  <span style="color:teal">Thursday</span> has the highest total vehicle count, with **23,997,463** vehicles, closely followed by <span style="color:teal">Tuesday</span> (**23,646,429**) and <span style="color:teal">Wednesday</span> (**23,626,935**). These three days represent the peak traffic days, indicating possibly higher commuter activity or business-related traffic.

- **<span style="color:teal">Friday Traffic Dip:</span>**  
  <span style="color:teal">Friday</span> sees a significant drop in traffic, with only **567,558** vehicles recorded. This low traffic count might be due to shorter working hours, reduced weekday commuter activity, or an anomaly in the dataset.

- **<span style="color:teal">Weekend Traffic:</span>**  
  <span style="color:teal">Saturday</span> has a moderate traffic count of **17,487,776**, suggesting that while traffic is lighter than during the week, there is still considerable movement. <span style="color:teal">Sunday</span> has the lowest traffic volume at **1,024,071**, which could be due to fewer commutes and business activities on that day.

### <span style="color:teal">Conclusion:</span>  
The highest traffic volumes are observed on weekdays, especially <span style="color:teal">Thursday</span>, while Fridays and Sundays show substantially lower vehicle counts. This indicates a typical weekday rush, with the weekend seeing reduced traffic.


##  Monthly Traffic Trends

```python
df["month"] = df["count_date"].dt.month_name()
monthly_traffic = df.groupby("month")["total_vehicles"].sum().reset_index()

# Order months correctly
month_order = [
    "January", "February", "March", "April", "May", "June", 
    "July", "August", "September", "October", "November", "December"
]
monthly_traffic["month"] = pd.Categorical(monthly_traffic["month"], categories=month_order, ordered=True)
monthly_traffic = monthly_traffic.sort_values("month")

print(monthly_traffic)
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_traffic, x="month", y="total_vehicles", marker="o", color="purple")
plt.xlabel("Month")
plt.ylabel("Total Traffic Volume")
plt.title("Traffic Volume Trends Across Different Months")
plt.xticks(rotation=45)
plt.show()
```

### <span style="color:teal">Monthly Traffic Trends:</span>

#### <span style="color:teal">Highest Traffic Volume:</span>
- **November** stands out as the month with the highest total vehicle count, reaching **12,632,846** vehicles, followed by **October** with **11,502,854** vehicles.

#### <span style="color:teal">Notable Trends:</span>
- **June** and **July** also show high traffic volumes with **10,853,248** and **3,237,282** vehicles, respectively.
- The months at the start of the year, particularly **January** and **February**, recorded the lowest traffic volumes, with **4,075,752** and **8,563,358** vehicles, respectively.

#### <span style="color:teal">Opportunities for Optimization:</span>
- The months with higher traffic volumes, especially **October**, **November**, and **June**, could benefit from focused traffic management strategies, such as optimizing traffic flow and introducing more sustainable transportation options to alleviate congestion.

#### <span style="color:teal">Seasonal Variations:</span>
- There is a notable peak in traffic during the later months of the year, suggesting that planning for holiday seasons or special events could be crucial for minimizing congestion during these times.
